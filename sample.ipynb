```python
import pandas as pd
import datetime as dt
import yfinance as yf
```

```python
earnings = pd.read_excel('data\earnings formatted.xlsx')
```

```python
aapl = yf.Ticker('AAPL').history(start=dt.datetime(2020,1,1))
aapl = aapl.reset_index()
nvda = yf.Ticker('NVDA').history(start=dt.datetime(2020,1,1))
nvda = nvda.reset_index()

aapl['Close shifted'] = aapl['Close'].shift(1)
aapl['-1d ret'] = aapl['Close']/aapl['Close shifted'] - 1
aapl['1d ret'] = aapl['-1d ret'].shift(-1)

nvda['Close shifted'] = nvda['Close'].shift(1)
nvda['-1d ret'] = nvda['Close']/nvda['Close shifted'] - 1
nvda['1d ret'] = nvda['-1d ret'].shift(-1)


returns = pd.concat([aapl,nvda],keys=['AAPL','NVDA']).reset_index()
returns = returns.rename(columns={'level_0':'Ticker'})
returns = returns.drop(columns=['Open', 'High', 'Low', 'Dividends', 'Stock Splits', 'level_1'])
returns
```

```python
earnings = earnings[earnings['Ticker'].isin(['AAPL','NVDA'])]
earnings
```

```python
merged = pd.merge(left=earnings, right=returns, how='left')
merged
```
